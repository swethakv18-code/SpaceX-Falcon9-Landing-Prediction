Proximity to the Equator

Proximity to the Coast

Coastal locations are chosen for safety and logistics: rockets launch eastward over the ocean, minimizing risk to populated areas and allowing debris to fall harmlessly into water.

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

def haversine(coord1, coord2):
    R = 6371.0  # Earth radius in km
    lat1, lon1 = coord1
    lat2, lon2 = coord2

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

launch_site = (28.563197, -80.576820)
coastline = (28.563, -80.567)

distance_coastline = haversine(launch_site, coastline)
print(f"Distance to coastline: {distance_coastline:.2f} km")

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

Close to infrastructure (railways, highways) for logistics.

Close to coastlines for safe launch trajectories.

Far enough from cities to reduce risk, but not so far that workforce and supply chains are disrupted.

This balance of proximity and distance reflects careful planning to maximize safety, efficiency, and accessibility.
